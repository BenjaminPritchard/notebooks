```python
from music21 import scale, stream, note

# Create a C major scale
c_major_scale = scale.MajorScale('C')

# Generate the pitches of the scale
pitches = c_major_scale.getPitches('C4', 'C6')  # From C4 to C5

# Create a stream (musical staff) and add the notes of the scale
staff = stream.Stream()
for pitch in pitches:
    n = note.Note(pitch)
    n.quarterLength = 0.25  # Set the duration to a 16th note
    staff.append(n)

# Show the staff (opens in a music notation viewer or renders as an image in Jupyter)
staff.show()
```

```python
from music21 import stream, note, clef, key, metadata

# Create the right hand (Treble clef) part
right_hand = stream.Part()
right_hand.append(clef.TrebleClef())
right_hand.append(key.KeySignature(2))  # D major has 2 sharps

d_major_scale = scale.MajorScale('D')
pitches = d_major_scale.getPitches('C4', 'C6')  # From C4 to C5
for pitch in pitches:
    n = note.Note(pitch)
    n.quarterLength = 0.25  # Set the duration to a 16th note
    right_hand.append(n)

# Create the left hand (Bass clef) part
left_hand = stream.Part()
left_hand.append(clef.BassClef())
left_hand.append(key.KeySignature(2))

# D major descending (octave lower)
d_major_desc = ['D3', 'C#3', 'B2', 'A2', 'G2', 'F#2', 'E2', 'D2']
for pitch in d_major_desc:
    left_hand.append(note.Note(pitch, quarterLength=1))

# Combine both hands into a score
score = stream.Score()

score.insert(0, right_hand)
score.insert(0, left_hand)

# Show as sheet music
score.show()
```
